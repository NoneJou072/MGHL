```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np

# 生成五组包含两个元素的数据
data = np.array([[0.96, 0.95, 0.05], [0.95, 0.9, 0.1], [0.9, 0.8, 0.0]])
label = ['HDTO', 'RHER', 'HER(TQC)']
# 设置画布大小
plt.figure(figsize=(5, 3))  # 设置宽度为6，高度为4

# 设置柱状图的宽度
bar_width = 0.15
index = np.arange(len(data))  # x轴位置

# 绘制柱状图
for i in range(len(data[0])):
    plt.bar(index + i * bar_width, data[:, i], width=bar_width, label=label[i])

# 设置图表标签、标题和图例
plt.ylabel('Success Rate', fontsize=12)
plt.ylim(0, 1.1)
plt.xticks(index + bar_width, ['DrawerBox', 'LockedCabinet', 'MultiCubes'], fontsize=12)
plt.tick_params(labelsize=12)
plt.legend(loc='lower center', bbox_to_anchor=(0.5, 1.0), ncol=5, fontsize=12)

plt.savefig('resources/images/full_task.png', bbox_inches='tight')

plt.show()
```
