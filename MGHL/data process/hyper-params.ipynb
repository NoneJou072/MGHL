## 评估子任务的成功率

```python
%matplotlib inline
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import os
```

# 导入 drawerbox 数据

```python
resource_dir = os.path.join(os.getcwd(), 'resources/hyper-eps')

drawer_001 = pd.read_csv(resource_dir + '/eps0.01-drawer.csv')
drawer_0015 = pd.read_csv(resource_dir + '/eps0.015-drawer.csv')
drawer_002 = pd.read_csv(resource_dir + '/eps0.02-drawer.csv')
drawer_0025 = pd.read_csv(resource_dir + '/eps0.025-drawer.csv')
drawer_003 = pd.read_csv(resource_dir + '/eps0.03-drawer.csv')

place_001 = pd.read_csv(resource_dir + '/eps0.01-place.csv')
place_0015 = pd.read_csv(resource_dir + '/eps0.015-place.csv')
place_002 = pd.read_csv(resource_dir + '/eps0.02-place.csv')
place_0025 = pd.read_csv(resource_dir + '/eps0.025-place.csv')
place_003 = pd.read_csv(resource_dir + '/eps0.03-place.csv')

drawer_001_val = drawer_001['Value']
place_001_val = place_001['Value']
drawer_001_timestep = 1e-5 * drawer_001['Step']

drawer_0015_val = drawer_0015['Value']
place_0015_val = place_0015['Value']
drawer_0015_timestep = 1e-5 * drawer_0015['Step']

drawer_002_val = drawer_002['Value']
place_002_val = place_002['Value']
drawer_002_timestep = 1e-5 * drawer_002['Step']

drawer_0025_val = drawer_0025['Value']
place_0025_val = place_0025['Value']
drawer_0025_timestep = 1e-5 * drawer_0025['Step']

drawer_003_val = drawer_003['Value']
place_003_val = place_003['Value']
drawer_003_timestep = 1e-5 * drawer_003['Step']

drawer_001_mean = drawer_001_val.rolling(30, min_periods=1).mean()
drawer_001_std = drawer_001_val.rolling(30, min_periods=1).std()

drawer_0015_mean = drawer_0015_val.rolling(30, min_periods=1).mean()
drawer_0015_std = drawer_0015_val.rolling(30, min_periods=1).std()

drawer_002_mean = drawer_002_val.rolling(30, min_periods=1).mean()
drawer_002_std = drawer_002_val.rolling(30, min_periods=1).std()

drawer_0025_mean = drawer_0025_val.rolling(30, min_periods=1).mean()
drawer_0025_std = drawer_0025_val.rolling(30, min_periods=1).std()

drawer_003_mean = drawer_003_val.rolling(30, min_periods=1).mean()
drawer_003_std = drawer_003_val.rolling(30, min_periods=1).std()

place_001_mean = place_001_val.rolling(30, min_periods=1).mean()
place_001_std = place_001_val.rolling(30, min_periods=1).std()

place_0015_mean = place_0015_val.rolling(30, min_periods=1).mean()
place_0015_std = place_0015_val.rolling(30, min_periods=1).std()

place_002_mean = place_002_val.rolling(30, min_periods=1).mean()
place_002_std = place_002_val.rolling(30, min_periods=1).std()

place_0025_mean = place_0025_val.rolling(30, min_periods=1).mean()
place_0025_std = place_0025_val.rolling(30, min_periods=1).std()

place_003_mean = place_003_val.rolling(30, min_periods=1).mean()
place_003_std = place_003_val.rolling(30, min_periods=1).std()
```

```python
# 创建一个包含两个子图的图形
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 3), dpi=80)

ax1.plot(drawer_001_timestep, drawer_001_mean, 'b', linewidth=2.0, label='0.01')
ax1.fill_between(drawer_001_timestep, drawer_001_mean - 0.5 * drawer_001_std, drawer_001_mean + 0.5 * drawer_001_std,color='b', alpha=0.2, linewidth=0)
ax1.plot(drawer_0015_timestep, drawer_0015_mean, 'g', linewidth=2.0, label='0.015')
ax1.fill_between(drawer_0015_timestep, drawer_0015_mean - 0.5 * drawer_0015_std, drawer_0015_mean + 0.5 * drawer_0015_std,color='g', alpha=0.2, linewidth=0)
ax1.plot(drawer_002_timestep, drawer_002_mean, 'r', linewidth=2.0, label='0.02')
ax1.fill_between(drawer_002_timestep, drawer_002_mean - 0.5 * drawer_002_std, drawer_002_mean + 0.5 * drawer_002_std,color='r', alpha=0.2, linewidth=0)
ax1.plot(drawer_0025_timestep, drawer_0025_mean, '#1f77b4', linewidth=2.0, label='0.025')
ax1.fill_between(drawer_0025_timestep, drawer_0025_mean - 0.5 * drawer_0025_std, drawer_0025_mean + 0.5 * drawer_0025_std,color='#1f77b4', alpha=0.2, linewidth=0)
ax1.plot(drawer_003_timestep, drawer_003_mean, '#ff7f0e', linewidth=2.0, label='0.03')
ax1.fill_between(drawer_003_timestep, drawer_003_mean - 0.5 * drawer_003_std, drawer_003_mean + 0.5 * drawer_003_std,color='#ff7f0e', alpha=0.2, linewidth=0)

ax1.set_xlim(0, 10)
ax1.set_ylim(-0.1, 1.1)
ax1.set_xlabel('Timesteps($\\times 10^5$)', fontsize=14)
ax1.set_ylabel('Success Rate', fontsize=14)
ax1.tick_params(axis='both', which='major', labelsize=14)
ax1.grid(linestyle='--')
ax1.text(0.1, 0.95, ' c', fontsize=16, color='black', fontweight='bold')
ax1.axhline(y=0.9, color='gray', linestyle='--', label='Success rate: 0.9')

ax2.plot(drawer_001_timestep, place_001_mean, 'b', linewidth=2.0, label='0.01')
ax2.fill_between(drawer_001_timestep, place_001_mean - 0.5 * place_001_std, place_001_mean + 0.5 * place_001_std,color='b', alpha=0.2, linewidth=0)
ax2.plot(drawer_001_timestep, place_0015_mean, 'g', linewidth=2.0, label='0.015')
ax2.fill_between(drawer_001_timestep, place_0015_mean - 0.5 * place_0015_std, place_0015_mean + 0.5 * place_0015_std,color='g', alpha=0.2, linewidth=0)
ax2.plot(drawer_001_timestep, place_002_mean, 'r', linewidth=2.0, label='0.02')
ax2.fill_between(drawer_001_timestep, place_002_mean - 0.5 * place_002_std, place_002_mean + 0.5 * place_002_std,color='r', alpha=0.2, linewidth=0)
ax2.plot(drawer_001_timestep, place_0025_mean, '#1f77b4', linewidth=2.0, label='0.025')
ax2.fill_between(drawer_001_timestep, place_0025_mean - 0.5 * place_0025_std, place_0025_mean + 0.5 * place_0025_std,color='#1f77b4', alpha=0.2, linewidth=0)
ax2.plot(drawer_001_timestep, place_003_mean, '#ff7f0e', linewidth=2.0, label='0.03')
ax2.fill_between(drawer_0015_timestep, place_003_mean - 0.5 * place_003_std, place_003_mean + 0.5 * place_003_std,color='#ff7f0e', alpha=0.2, linewidth=0)

ax2.set_xlim(0, 10)
ax2.set_ylim(-0.1, 1.1)
ax2.set_xlabel('Timesteps($\\times 10^5$)', fontsize=14)
ax2.set_ylabel('Success Rate', fontsize=14)
ax2.tick_params(axis='both', which='major', labelsize=14)
ax2.grid(linestyle='--')
ax2.text(0.1, 0.95, ' d', fontsize=16, color='black', fontweight='bold')
ax2.axhline(y=0.9, color='gray', linestyle='--', label='Success rate: 0.9')

# 创建共享的图例
handles, labels = ax1.get_legend_handles_labels()
fig.legend(handles, labels, loc='lower center', bbox_to_anchor=(0.5, 1), ncol=6, fontsize=14)

# 调整子图之间的间距
plt.tight_layout()

# 保存图形
plt.savefig('resources/images/combined_plot.png', bbox_inches='tight')

# 显示图形
plt.show()
```
