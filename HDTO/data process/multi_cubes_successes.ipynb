## 评估子任务的成功率

```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import os
```

```python
resource_dir = os.path.join(os.getcwd(), 'resources')
blue_sac_success_rate_frame = pd.read_csv(resource_dir + '/log_runs_IORL_MultiCubeStack-v1_num_1_seed_10_2_1.csv')
green_sac_success_rate_frame = pd.read_csv(resource_dir + '/log_runs_IORL_MultiCubeStack-v1_num_1_seed_10_2_2.csv')
red_sac_success_rate_frame = pd.read_csv(resource_dir + '/log_runs_IORL_MultiCubeStack-v1_num_1_seed_10_2_3.csv')

blue_sac_success_rate = blue_sac_success_rate_frame['Value']
green_sac_success_rate = green_sac_success_rate_frame['Value']
red_sac_success_rate = red_sac_success_rate_frame['Value']

timesteps_sac = blue_sac_success_rate_frame['Step']
```

```python
red_sac_success_rate_mean = red_sac_success_rate.rolling(5).mean()
red_sac_success_rate_std = red_sac_success_rate.rolling(5).std()

plt.figure()
plt.plot(timesteps_sac, red_sac_success_rate_mean, 'r', label='SAC')
plt.fill_between(timesteps_sac, red_sac_success_rate_mean - 2 * red_sac_success_rate_std, red_sac_success_rate_mean + 2 * red_sac_success_rate_std, color='b', alpha=0.2)

# set axis range
# plt.xlim(0, 1.5e5)
plt.ylim(-0.1, 1.1)

# set axis label
plt.xlabel('timesteps', fontsize=12)
plt.ylabel('success rate', fontsize=12)

# display legend
plt.legend(loc='lower right', fontsize=12)
plt.title('Stack-Success-Rate-Red')
# save figure
plt.savefig('resources/images/stack-success-rate-red.png')

plt.show()
```

```python
green_sac_success_rate_mean = green_sac_success_rate.rolling(5).mean()
green_sac_success_rate_std = green_sac_success_rate.rolling(5).std()

plt.figure()
plt.plot(timesteps_sac, green_sac_success_rate_mean, 'r', label='SAC')
plt.fill_between(timesteps_sac, green_sac_success_rate_mean - 2 * green_sac_success_rate_std, green_sac_success_rate_mean + 2 * green_sac_success_rate_std, color='b', alpha=0.2)

# set axis range
# plt.xlim(0, 1.5e5)
plt.ylim(-0.1, 1.1)

# set axis label
plt.xlabel('timesteps', fontsize=12)
plt.ylabel('success rate', fontsize=12)

# display legend
plt.legend(loc='lower right', fontsize=12)
plt.title('Stack-Success-Rate-Green')
# save figure
plt.savefig('resources/images/stack-success-rate-green.png')

plt.show()
```

```python
blue_sac_success_rate_mean = blue_sac_success_rate.rolling(5).mean()
blue_sac_success_rate_std = blue_sac_success_rate.rolling(5).std()

plt.figure()
plt.plot(timesteps_sac, blue_sac_success_rate_mean, 'r', label='SAC')
plt.fill_between(timesteps_sac, blue_sac_success_rate_mean - 2 * blue_sac_success_rate_std, blue_sac_success_rate_mean + 2 * blue_sac_success_rate_std, color='b', alpha=0.2)

# set axis range
# plt.xlim(0, 1.5e5)
plt.ylim(-0.1, 1.1)

# set axis label
plt.xlabel('timesteps', fontsize=12)
plt.ylabel('success rate', fontsize=12)

# display legend
plt.legend(loc='lower right', fontsize=12)
plt.title('Stack-Success-Rate-Blue')
# save figure
plt.savefig('resources/images/stack-success-rate-blue.png')

plt.show()
```
