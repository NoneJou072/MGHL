## 评估子任务的成功率

```python
%matplotlib inline
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import os
```

```python
resource_dir = os.path.join(os.getcwd(), 'resources/cubes_data/')

blue_sac_success_rate_frame = pd.read_csv(resource_dir + '/HDTO_MultiCubeStack-v1_blue.csv')
green_sac_success_rate_frame = pd.read_csv(resource_dir + '/HDTO_MultiCubeStack-v1_green.csv')
red_sac_success_rate_frame = pd.read_csv(resource_dir + '/HDTO_MultiCubeStack-v1_red.csv')

blue_td3_success_rate_frame = pd.read_csv(resource_dir + '/log_runs_IORL_DrawerBox-v1_num_2_seed_10_2.csv')
green_td3_success_rate_frame = pd.read_csv(resource_dir + '/log_runs_IORL_DrawerBox-v1_num_2_seed_10_2_1.csv')
red_td3_success_rate_frame = pd.read_csv(resource_dir + '/log_runs_IORL_DrawerBox-v1_num_2_seed_10_2_2.csv')

blue_her_success_rate_frame = pd.read_csv(resource_dir + '/log_runs_IORL_MultiCubeStack-v1_num_3_seed_10.csv')
green_her_success_rate_frame = pd.read_csv(resource_dir + '/log_runs_IORL_MultiCubeStack-v1_num_3_seed_10_2.csv')
red_her_success_rate_frame = pd.read_csv(resource_dir + '/log_runs_IORL_MultiCubeStack-v1_num_3_seed_10_3.csv')

blue_sac_success_rate = blue_sac_success_rate_frame['Value']
green_sac_success_rate = green_sac_success_rate_frame['Value']
red_sac_success_rate = red_sac_success_rate_frame['Value']
timesteps_sac = 1e-5 * blue_sac_success_rate_frame['Step']

blue_td3_success_rate = blue_td3_success_rate_frame['Value']
green_td3_success_rate = green_td3_success_rate_frame['Value']
red_td3_success_rate = red_td3_success_rate_frame['Value']
timesteps_td3 = 1e-5 * blue_td3_success_rate_frame['Step']

blue_her_success_rate = blue_her_success_rate_frame['Value']
green_her_success_rate = green_her_success_rate_frame['Value']
red_her_success_rate = red_her_success_rate_frame['Value']
timesteps_her = 1e-5 * blue_her_success_rate_frame['Step']

blue_sac_success_rate_mean = blue_sac_success_rate.rolling(30).mean()
blue_sac_success_rate_std = blue_sac_success_rate.rolling(30).std()

blue_td3_success_rate_mean = blue_td3_success_rate.rolling(30).mean()
blue_td3_success_rate_std = blue_td3_success_rate.rolling(30).std()

blue_her_success_rate_mean = blue_her_success_rate.rolling(30).mean()
blue_her_success_rate_std = blue_her_success_rate.rolling(30).std()

green_sac_success_rate_mean = green_sac_success_rate.rolling(30).mean()
green_sac_success_rate_std = green_sac_success_rate.rolling(30).std()

green_td3_success_rate_mean = green_td3_success_rate.rolling(30).mean()
green_td3_success_rate_std = green_td3_success_rate.rolling(30).std()

green_her_success_rate_mean = green_her_success_rate.rolling(30).mean()
green_her_success_rate_std = green_her_success_rate.rolling(30).std()

red_sac_success_rate_mean = red_sac_success_rate.rolling(30).mean()
red_sac_success_rate_std = red_sac_success_rate.rolling(30).std()

red_td3_success_rate_mean = red_td3_success_rate.rolling(30).mean()
red_td3_success_rate_std = red_td3_success_rate.rolling(30).std()

red_her_success_rate_mean = red_her_success_rate.rolling(30).mean()
red_her_success_rate_std = red_her_success_rate.rolling(30).std()
```

```python
fig, (ax1, ax2, ax3) = plt.subplots(1, 3, figsize=(12, 3), dpi=80)

# ax1
ax1.plot(timesteps_sac, blue_sac_success_rate_mean, 'b', linewidth=2.0, label='HGHL')
ax1.fill_between(timesteps_sac, blue_sac_success_rate_mean - 0.5 * blue_sac_success_rate_std, blue_sac_success_rate_mean + 0.5 * blue_sac_success_rate_std,color='b', alpha=0.2, linewidth=0)
ax1.plot(timesteps_td3, blue_td3_success_rate_mean, 'g', linewidth=2.0, label='HGHL(TD3)')
ax1.fill_between(timesteps_td3, blue_td3_success_rate_mean - 0.5 * blue_td3_success_rate_std, blue_td3_success_rate_mean + 0.5 * blue_td3_success_rate_std,color='g', alpha=0.2, linewidth=0)
ax1.plot(timesteps_her, blue_her_success_rate_mean, 'y', linewidth=2.0, label='HER(SAC)')
ax1.fill_between(timesteps_her, blue_her_success_rate_mean - 0.5 * blue_her_success_rate_std, blue_her_success_rate_mean + 0.5 * blue_her_success_rate_std,color='y', alpha=0.2, linewidth=0)

ax1.set_xlim(0, 17)
ax1.set_ylim(-0.1, 1.1)
ax1.set_xlabel('Timesteps($\\times 10^5$)', fontsize=14)
ax1.set_ylabel('Success Rate', fontsize=14)
ax1.tick_params(axis='both', which='major', labelsize=14)
ax1.grid(linestyle='--')
ax1.text(0.1, 0.95, ' a', fontsize=16, color='black', fontweight='bold')

# ax2
ax2.plot(timesteps_sac, green_sac_success_rate_mean, 'b', linewidth=2.0, label='HDTO')
ax2.fill_between(timesteps_sac, green_sac_success_rate_mean - 0.5 * green_sac_success_rate_std, green_sac_success_rate_mean + 0.5 * green_sac_success_rate_std,color='b', alpha=0.2, linewidth=0)
ax2.plot(timesteps_td3, green_td3_success_rate_mean, 'g', linewidth=2.0, label='HDTOxTD3')
ax2.fill_between(timesteps_td3, green_td3_success_rate_mean - 0.5 * green_td3_success_rate_std, green_td3_success_rate_mean + 0.5 * green_td3_success_rate_std,color='g', alpha=0.2, linewidth=0)
ax2.plot(timesteps_her, green_her_success_rate_mean, 'y', linewidth=2.0, label='HERxSAC')
ax2.fill_between(timesteps_her, green_her_success_rate_mean - 0.5 * green_her_success_rate_std, green_her_success_rate_mean + 0.5 * green_her_success_rate_std,color='y', alpha=0.2, linewidth=0)

ax2.set_xlim(0, 17)
ax2.set_ylim(-0.1, 1.1)
ax2.set_xlabel('Timesteps($\\times 10^5$)', fontsize=14)
ax2.set_ylabel('Success Rate', fontsize=14)
ax2.tick_params(axis='both', which='major', labelsize=14)
ax2.grid(linestyle='--')
ax2.text(0.1, 0.95, ' b', fontsize=16, color='black', fontweight='bold')

# ax3
ax3.plot(timesteps_sac, red_sac_success_rate_mean, 'b', linewidth=2.0, label='HDTO')
ax3.fill_between(timesteps_sac, red_sac_success_rate_mean - 0.5 * red_sac_success_rate_std, red_sac_success_rate_mean + 0.5 * red_sac_success_rate_std,color='b', alpha=0.2, linewidth=0)
ax3.plot(timesteps_td3, red_td3_success_rate_mean, 'g', linewidth=2.0, label='HDTOxTD3')
ax3.fill_between(timesteps_td3, red_td3_success_rate_mean - 0.5 * red_td3_success_rate_std, red_td3_success_rate_mean + 0.5 * red_td3_success_rate_std,color='g', alpha=0.2, linewidth=0)
ax3.plot(timesteps_her, red_her_success_rate_mean, 'y', linewidth=2.0, label='HERxSAC')
ax3.fill_between(timesteps_her, red_her_success_rate_mean - 0.5 * red_her_success_rate_std, red_her_success_rate_mean + 0.5 * red_her_success_rate_std,color='y', alpha=0.2, linewidth=0)

ax3.set_xlim(0, 17)
ax3.set_ylim(-0.1, 1.1)
ax3.set_xlabel('Timesteps($\\times 10^5$)', fontsize=14)
ax3.set_ylabel('Success Rate', fontsize=14)
ax3.tick_params(axis='both', which='major', labelsize=14)
ax3.grid(linestyle='--')
ax3.text(0.1, 0.95, ' c', fontsize=16, color='black', fontweight='bold')

# ax1 - Add vertical dashed line at y=0.8
ax1.axhline(y=0.8, color='gray', linestyle='--', label='Success rate: 0.8')
ax1.axhline(y=0.9, color='black', linestyle='--', label='Success rate: 0.9')

# ax2 - Add vertical dashed line at y=0.8
ax2.axhline(y=0.8, color='gray', linestyle='--', label='0.8')
ax2.axhline(y=0.9, color='black', linestyle='--', label='0.9')

# ax3 - Add vertical dashed line at y=0.8
ax3.axhline(y=0.8, color='gray', linestyle='--', label='0.8')
ax3.axhline(y=0.9, color='black', linestyle='--', label='0.9')

# 创建共享的图例
handles, labels = ax1.get_legend_handles_labels()
fig.legend(handles, labels, loc='lower center', bbox_to_anchor=(0.5, 1), ncol=5, fontsize=16)

plt.tight_layout()

plt.savefig('resources/images/cubes-eva.png', bbox_inches='tight')

plt.show()
```
