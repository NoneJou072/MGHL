## 评估子任务的成功率

```python
%matplotlib inline
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import os
```

```python
resource_dir = os.path.join(os.getcwd(), 'resources/cubes_data/')

blue_sac_success_rate_frame = pd.read_csv(resource_dir + '/HDTO_MultiCubeStack-v1_num_3_seed_10_1.csv')
green_sac_success_rate_frame = pd.read_csv(resource_dir + '/HDTO_MultiCubeStack-v1_num_3_seed_10_1_2.csv')
red_sac_success_rate_frame = pd.read_csv(resource_dir + '/HDTO_MultiCubeStack-v1_num_3_seed_10_1_3.csv')

blue_td3_success_rate_frame = pd.read_csv(resource_dir + '/log_runs_IORL_DrawerBox-v1_num_2_seed_10_2.csv')
green_td3_success_rate_frame = pd.read_csv(resource_dir + '/log_runs_IORL_DrawerBox-v1_num_2_seed_10_2_1.csv')
red_td3_success_rate_frame = pd.read_csv(resource_dir + '/log_runs_IORL_DrawerBox-v1_num_2_seed_10_2_2.csv')

blue_her_success_rate_frame = pd.read_csv(resource_dir + '/log_runs_IORL_MultiCubeStack-v1_num_3_seed_10.csv')
green_her_success_rate_frame = pd.read_csv(resource_dir + '/log_runs_IORL_MultiCubeStack-v1_num_3_seed_10_2.csv')
red_her_success_rate_frame = pd.read_csv(resource_dir + '/log_runs_IORL_MultiCubeStack-v1_num_3_seed_10_3.csv')

blue_sac_success_rate = blue_sac_success_rate_frame['Value']
green_sac_success_rate = green_sac_success_rate_frame['Value']
red_sac_success_rate = red_sac_success_rate_frame['Value']
timesteps_sac = blue_sac_success_rate_frame['Step']

blue_td3_success_rate = blue_td3_success_rate_frame['Value']
green_td3_success_rate = green_td3_success_rate_frame['Value']
red_td3_success_rate = red_td3_success_rate_frame['Value']
timesteps_td3 = blue_td3_success_rate_frame['Step']

blue_her_success_rate = blue_her_success_rate_frame['Value']
green_her_success_rate = green_her_success_rate_frame['Value']
red_her_success_rate = red_her_success_rate_frame['Value']
timesteps_her = blue_her_success_rate_frame['Step']
```

```python
blue_sac_success_rate_mean = blue_sac_success_rate.rolling(10).mean()
blue_sac_success_rate_std = blue_sac_success_rate.rolling(10).std()

blue_td3_success_rate_mean = blue_td3_success_rate.rolling(10).mean()
blue_td3_success_rate_std = blue_td3_success_rate.rolling(10).std()

blue_her_success_rate_mean = blue_her_success_rate.rolling(10).mean()
blue_her_success_rate_std = blue_her_success_rate.rolling(10).std()

plt.figure(dpi=100,figsize=(4,3))
plt.plot(timesteps_sac, blue_sac_success_rate_mean, 'b', linewidth=2.0, label='HDTO')
plt.fill_between(timesteps_sac, blue_sac_success_rate_mean - 0.5 * blue_sac_success_rate_std, blue_sac_success_rate_mean + 0.5 * blue_sac_success_rate_std,color='b', alpha=0.2)
plt.plot(timesteps_td3, blue_td3_success_rate_mean, 'g', linewidth=2.0, label='HDTOxTD3')
plt.fill_between(timesteps_td3, blue_td3_success_rate_mean - 0.5 * blue_td3_success_rate_std, blue_td3_success_rate_mean + 0.5 * blue_td3_success_rate_std,color='g', alpha=0.2)
plt.plot(timesteps_her, blue_her_success_rate_mean, 'y', linewidth=2.0, label='HERxSAC')
plt.fill_between(timesteps_her, blue_her_success_rate_mean - 0.5 * blue_her_success_rate_std, blue_her_success_rate_mean + 0.5 * blue_her_success_rate_std,color='y', alpha=0.2)

# set axis range
plt.xlim(0, 1.7e6)
plt.ylim(-0.1, 1.1)

# set axis label
plt.xlabel('timesteps', fontsize=12)
plt.ylabel('success rate', fontsize=12)
plt.ticklabel_format(style='sci', axis='x', scilimits=(0,0))
# plt.tick_params(axis='both', labelsize=14)
# X、Y轴刻度标签字体大小
matplotlib.rcParams['axes.labelsize'] = 14

plt.legend(loc='lower center', bbox_to_anchor=(0.5, -0.4), ncol=5, fontsize=12)

plt.title('MultiCube-Blue-Success-Rate', fontsize=14)
# save figure
plt.savefig('resources/images/cubes-blue-rate.png', bbox_inches='tight')

plt.show()
```

```python
green_sac_success_rate_mean = green_sac_success_rate.rolling(10).mean()
green_sac_success_rate_std = green_sac_success_rate.rolling(10).std()

green_td3_success_rate_mean = green_td3_success_rate.rolling(10).mean()
green_td3_success_rate_std = green_td3_success_rate.rolling(10).std()

green_her_success_rate_mean = green_her_success_rate.rolling(10).mean()
green_her_success_rate_std = green_her_success_rate.rolling(10).std()

plt.figure(dpi=100,figsize=(4,3))
plt.plot(timesteps_sac, green_sac_success_rate_mean, 'b', linewidth=2.0, label='HDTO')
plt.fill_between(timesteps_sac, green_sac_success_rate_mean - 0.5 * green_sac_success_rate_std, green_sac_success_rate_mean + 0.5 * green_sac_success_rate_std,color='b', alpha=0.2)
plt.plot(timesteps_td3, green_td3_success_rate_mean, 'g', linewidth=2.0, label='HDTOxTD3')
plt.fill_between(timesteps_td3, green_td3_success_rate_mean - 0.5 * green_td3_success_rate_std, green_td3_success_rate_mean + 0.5 * green_td3_success_rate_std,color='g', alpha=0.2)
plt.plot(timesteps_her, green_her_success_rate_mean, 'y', linewidth=2.0, label='HERxSAC')
plt.fill_between(timesteps_her, green_her_success_rate_mean - 0.5 * green_her_success_rate_std, green_her_success_rate_mean + 0.5 * green_her_success_rate_std,color='y', alpha=0.2)

# set axis range
plt.xlim(0, 1.7e6)
plt.ylim(-0.1, 1.1)

# set axis label
plt.xlabel('timesteps', fontsize=12)
plt.ylabel('success rate', fontsize=12)
plt.ticklabel_format(style='sci', axis='x', scilimits=(0,0))
# plt.tick_params(axis='both', labelsize=14)
# X、Y轴刻度标签字体大小
matplotlib.rcParams['axes.labelsize'] = 14

plt.legend(loc='lower center', bbox_to_anchor=(0.5, -0.4), ncol=5, fontsize=12)

plt.title('MultiCube-Green-Success-Rate', fontsize=14)
# save figure
plt.savefig('resources/images/cubes-green-rate.png', bbox_inches='tight')

plt.show()
```
