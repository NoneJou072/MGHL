## 评估子任务的成功率

```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import os
```

```python
resource_dir = os.path.join(os.getcwd(), 'resources')
drawerbox_drawer_sac_success_rate_frame = pd.read_csv(resource_dir + '/log_runs_IORL_LockedCabinet-v1_num_1_seed_10_3_unlock.csv')
drawerbox_place_sac_success_rate_frame = pd.read_csv(resource_dir + '/log_runs_IORL_LockedCabinet-v1_num_1_seed_10_3_door.csv')

drawerbox_drawer_sac_success_rate = drawerbox_drawer_sac_success_rate_frame['Value']
drawerbox_place_sac_success_rate = drawerbox_place_sac_success_rate_frame['Value']
timesteps_sac = drawerbox_drawer_sac_success_rate_frame['Step']
```

```python
drawerbox_drawer_sac_success_rate_mean = drawerbox_drawer_sac_success_rate.rolling(5).mean()
drawerbox_drawer_sac_success_rate_std = drawerbox_drawer_sac_success_rate.rolling(5).std()

plt.figure()
plt.plot(timesteps_sac, drawerbox_drawer_sac_success_rate_mean, 'b', label='SAC')
plt.fill_between(timesteps_sac, drawerbox_drawer_sac_success_rate_mean - 2 * drawerbox_drawer_sac_success_rate_std, drawerbox_drawer_sac_success_rate_mean + 2 * drawerbox_drawer_sac_success_rate_std,color='b', alpha=0.2)

# set axis range
plt.xlim(0, 1.5e5)
plt.ylim(-0.1, 1.1)

# set axis label
plt.xlabel('timesteps', fontsize=12)
plt.ylabel('success rate', fontsize=12)

# display legend
plt.legend(loc='lower right', fontsize=12)
plt.title('LockedCabinet-Success-Rate-(Unlock Stage)')
# save figure
plt.savefig('resources/images/LockedCabinet-Success-Rate-(Unlock Stage).png')

plt.show()
```

```python
drawerbox_place_sac_success_rate_mean = drawerbox_place_sac_success_rate.rolling(5).mean()
drawerbox_place_sac_success_rate_std = drawerbox_place_sac_success_rate.rolling(5).std()

plt.figure()
plt.plot(timesteps_sac, drawerbox_place_sac_success_rate_mean, 'b', label='SAC')
plt.fill_between(timesteps_sac, drawerbox_place_sac_success_rate_mean - 2 * drawerbox_place_sac_success_rate_std, drawerbox_place_sac_success_rate_mean + 2 * drawerbox_place_sac_success_rate_std,color='b', alpha=0.2)

# set axis range
plt.xlim(0, 1.5e5)
plt.ylim(-0.1, 1.1)

# set axis label
plt.xlabel('timesteps', fontsize=12)
plt.ylabel('success rate', fontsize=12)

# display legend
plt.legend(loc='lower right', fontsize=12)
plt.title('LockedCabinet-Success-Rate-(Door Stage)')
# save figure
plt.savefig('resources/images/LockedCabinet-Success-Rate-(Door Stage).png')

plt.show()
```
