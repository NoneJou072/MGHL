## 评估子任务的成功率

```python
%matplotlib inline
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import os
```

```python
resource_dir = os.path.join(os.getcwd(), 'resources/drawer_data')
drawerbox_drawer_sac_success_rate_frame = pd.read_csv(resource_dir + '/log_runs_IORL_DrawerBox-v1_num_4_seed_10.csv')
drawerbox_drawer_td3_success_rate_frame = pd.read_csv(resource_dir + '/log_runs_IMPTD3_DrawerBox-v1_num_3_seed_10_1.csv')
drawerbox_drawer_her_success_rate_frame = pd.read_csv(resource_dir + '/log_runs_her_DrawerBox-v1_num_4_seed_10.csv')
drawerbox_place_sac_success_rate_frame = pd.read_csv(resource_dir + '/log_runs_IORL_DrawerBox-v1_num_4_seed_10_1.csv')
drawerbox_place_td3_success_rate_frame = pd.read_csv(resource_dir + '/log_runs_IMPTD3_DrawerBox-v1_num_3_seed_10_1_1.csv')
drawerbox_place_her_success_rate_frame = pd.read_csv(resource_dir + '/log_runs_her_DrawerBox-v1_num_4_seed_10_1.csv')

drawerbox_drawer_sac_success_rate = drawerbox_drawer_sac_success_rate_frame['Value']
drawerbox_place_sac_success_rate = drawerbox_place_sac_success_rate_frame['Value']
timesteps_sac = drawerbox_drawer_sac_success_rate_frame['Step']

drawerbox_drawer_td3_success_rate = drawerbox_drawer_td3_success_rate_frame['Value']
drawerbox_place_td3_success_rate = drawerbox_place_td3_success_rate_frame['Value']
timesteps_td3 = drawerbox_drawer_td3_success_rate_frame['Step']

drawerbox_drawer_her_success_rate = drawerbox_drawer_her_success_rate_frame['Value']
drawerbox_place_her_success_rate = drawerbox_place_her_success_rate_frame['Value']
timesteps_her = drawerbox_drawer_her_success_rate_frame['Step']
```

```python
drawerbox_drawer_sac_success_rate_mean = drawerbox_drawer_sac_success_rate.rolling(10).mean()
drawerbox_drawer_sac_success_rate_std = drawerbox_drawer_sac_success_rate.rolling(10).std()

drawerbox_drawer_td3_success_rate_mean = drawerbox_drawer_td3_success_rate.rolling(10).mean()
drawerbox_drawer_td3_success_rate_std = drawerbox_drawer_td3_success_rate.rolling(10).std()

drawerbox_drawer_her_success_rate_mean = drawerbox_drawer_her_success_rate.rolling(10).mean()
drawerbox_drawer_her_success_rate_std = drawerbox_drawer_her_success_rate.rolling(10).std()

plt.figure(dpi=100,figsize=(6,3))
plt.plot(timesteps_sac, drawerbox_drawer_sac_success_rate_mean, 'b', linewidth=2.0, label='HDTO')
plt.fill_between(timesteps_sac, drawerbox_drawer_sac_success_rate_mean - 0.5 * drawerbox_drawer_sac_success_rate_std, drawerbox_drawer_sac_success_rate_mean + 0.5 * drawerbox_drawer_sac_success_rate_std,color='b', alpha=0.2)
plt.plot(timesteps_td3, drawerbox_drawer_td3_success_rate_mean, 'g', linewidth=2.0, label='HDTOxTD3')
plt.fill_between(timesteps_td3, drawerbox_drawer_td3_success_rate_mean - 0.5 * drawerbox_drawer_td3_success_rate_std, drawerbox_drawer_td3_success_rate_mean + 0.5 * drawerbox_drawer_td3_success_rate_std,color='g', alpha=0.2)
plt.plot(timesteps_her, drawerbox_drawer_her_success_rate_mean, 'y', linewidth=2.0, label='HERxSAC')
plt.fill_between(timesteps_her, drawerbox_drawer_her_success_rate_mean - 0.5 * drawerbox_drawer_her_success_rate_std, drawerbox_drawer_her_success_rate_mean + 0.5 * drawerbox_drawer_her_success_rate_std,color='y', alpha=0.2)

# set axis range
plt.xlim(0, 3.5e5)
plt.ylim(-0.1, 1.1)

# set axis label
plt.xlabel('timesteps', fontsize=12)
plt.ylabel('success rate', fontsize=12)
plt.ticklabel_format(style='sci', axis='x', scilimits=(0,0))
# plt.tick_params(axis='both', labelsize=14)
# X、Y轴刻度标签字体大小
matplotlib.rcParams['axes.labelsize'] = 14

plt.legend(loc='lower center', bbox_to_anchor=(0.5, -0.4), ncol=5, fontsize=12)

plt.title('DrawerBox-Drawer-Success-Rate', fontsize=14)
# save figure
plt.savefig('resources/images/DrawerBox-Success-Rate-Drawer.png', bbox_inches='tight')

plt.show()
```
