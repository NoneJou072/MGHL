```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np

# 生成五组包含两个元素的数据
data = np.array([[0.85, 0.6, 0.5, 0.8, 0.2], [0.8, 0.3, 0.2, 0.01, 0.01],
                [0.9, 0.6, 0.8, 0.3, 0.2], [0.9, 0.3, 0.2, 0.01, 0.01]])
label = ['HDTO', 'HDTOxTD3', 'HERxSAC', 'RHER', 'HERxTQC']
# 设置画布大小
plt.figure(figsize=(6, 3))  # 设置宽度为6，高度为4

# 设置柱状图的宽度
bar_width = 0.15
index = np.arange(len(data))  # x轴位置

# 绘制柱状图
for i in range(len(data[0])):
    plt.bar(index + i * bar_width, data[:, i], width=bar_width, label=label[i])

# 设置图表标签、标题和图例
plt.ylabel('success rate')
plt.title('Full Task Evaluation')
plt.xticks(index + bar_width, ['DrawerBox', 'LockedCabinet', 'MultiCubes', 'Task 4'])
plt.legend(loc='lower center', bbox_to_anchor=(0.5, -0.3), ncol=5)

plt.ylim(0, 1.0)  # 设置y轴范围为0到7

plt.savefig('resources/images/full_task-success.png', bbox_inches='tight')
# 显示图表
plt.show()
```
